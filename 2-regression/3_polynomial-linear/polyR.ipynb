##### Import Dataset & Preprocess

```python
import pandas as pd
import numpy as np
dataset = pd.read_csv('./Position_Salaries.csv')
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(x)
print(y)
```

##### Training Models to Compare

###### Simple Linear Model

```python
from sklearn.linear_model import LinearRegression
LReg = LinearRegression()
LReg.fit(x, y)
```

###### Polynomial Linear Model

```python
#first create feature matrix of x, x^2, x^3...... x^n
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
x_poly = poly_reg.fit_transform(x) #returns new matrix of features composed of polynoms of x
LRegPoly = LinearRegression()
LRegPoly.fit(x_poly, y)
```

```python
print(x_poly)
```

##### Visualising Models

```python
import matplotlib.pyplot as plt
```

###### Simple Model

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, LReg.predict(x), color = 'blue')
plt.title('Estimator(Simple Model)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

###### Polynomial Model

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, LRegPoly.predict(x_poly), color = 'blue') #need to use the polynomial feature set for this
plt.title('Estimator(Poly Model)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

###### Poly Model in for smoother curve

```python
x_grid = np.arange(min(x), max(x), 0.1)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, LRegPoly.predict(poly_reg.fit_transform(x_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

##### Predictions

###### Simple Model

```python
print(LReg.predict([[6.5]]))
```

###### Polynomial Model

```python
print(LRegPoly.predict(poly_reg.fit_transform([[6.5]])))
```
